```python
import json
from tqdm import tqdm
import re
```

```python
text_template = "../data/template_english.txt"
```

```python
with open(text_template, 'r', encoding='utf-8') as f:
    lines = f.readlines()
    total_lines = len(lines)
    final_json = []
    
    i = 0
    move, step, choice = 0,0,0
    while i<total_lines:
        # print('outer loop')
        line = lines[i].strip()
        if not line:
            i+=1
            continue
            
        # check if line is a move.step
        if re.match(r'^\d+\.\d+$', line):
            move, step = line.split(".")
            # increment to get the choice
        elif re.match(r'^[A-Z]$', line):
            choice = line
        elif line.startswith('{'):
            # start tracking text
            curr_line = ''
            j = i
            while True:
                line_t = lines[j].strip()
                curr_line+= "\n"+line_t.lstrip('{').rstrip('}')
                j+=1
                if j>=total_lines or line_t.endswith('}'):
                    break
            text = curr_line.strip()
            
            # assign sub-choice
            sub_choice = 1
            if final_json:
                if final_json[-1]['move']==move and final_json[-1]['step']==step \
                    and final_json[-1]['choice']==choice:
                    prev_sub_choice = final_json[-1]['sub_choice']
                    sub_choice = prev_sub_choice + 1
                    
            final_json.append({"move": move, "step": step, "choice": choice, "sub_choice": sub_choice, "text": text})
            i = j-1 # i incremented at end always
        elif len(line)>1:
            # assign sub-choice
            sub_choice = 1
            if final_json:
                if final_json[-1]['move']==move and final_json[-1]['step']==step \
                    and final_json[-1]['choice']==choice:
                    prev_sub_choice = final_json[-1]['sub_choice']
                    sub_choice = prev_sub_choice + 1
                    
            text = line
            final_json.append({"move": move, "step": step, "choice": choice, "sub_choice": sub_choice, "text": text})
        i+=1
```

```python
for js in final_json:
    print(js)
    print('\n')
```

```python
write_file = '../data/template_english_dict.jsonl'
with open(write_file, 'w') as f:
    for line in final_json:
        json.dump(line, f)
        f.write('\n')
```
